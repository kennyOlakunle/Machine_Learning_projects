# Import packages

```python
import warnings

warnings.filterwarnings('ignore') # To ignore warnings
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression


%matplotlib inline
sns.set_style('whitegrid')
```

# Set warnings off and set seaborn

# Read and load data

```python
# Read the data

rev_data = pd.read_csv('cost_revenue_clean.csv')

rev_data.shape
```

```python
rev_data.head()
```

# Data Exploration & Visualisation

```python
#seperate the data into dependent and independent variables

X = pd.DataFrame(rev_data, columns=['production_budget_usd']) #independent variable/feature

y = pd.DataFrame(rev_data, columns=['worldwide_gross_usd']) # dependent variable/target
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

# Linear Regression

```python
regression = LinearRegression()
regression.fit(X, y)
```

### Slope coefficient

```python
#The slope coefficient is the average change in the response variable for every one unit increase in predictor variable X.
```

```python
# Slope Coefficient
regression.coef_ # theta_1
```

The result above shows that there is a positive relationship between budget and revenue, and it means that for each dollar spent, we should get like 3.1 dollars in return.

```python
regression.intercept_ # theta_0
```

Intercept - This is telling us that the movie with a budget of 0 - zero will loose over 7 million dollars.

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=4) # regression line
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#check the accuracy of the model

regression.score(X, y) # R^2 value

#
```
